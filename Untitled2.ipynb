```python
import streamlit as st

# Page setup
st.set_page_config(page_title="My Portfolio", layout="centered")

# Portfolio header
st.title("üßë‚Äçüíª Personal Portfolio")
st.subheader("Welcome to my portfolio app!")

# User inputs
name = st.text_input("Enter your name", placeholder="e.g. Alex Johnson")
role = st.text_input("Your role", placeholder="e.g. Software Developer")
about = st.text_area("About you", placeholder="Write a short bio...")
email = st.text_input("Email")
github = st.text_input("GitHub URL")
linkedin = st.text_input("LinkedIn URL")

# Generate portfolio view
if st.button("Generate Portfolio"):
    if not name:
        st.warning("Please enter your name.")
    else:
        st.success("Portfolio generated successfully!")

        # Display portfolio directly with Streamlit components
        st.header(name)
        st.subheader(role)

        st.write("### About Me")
        st.write(about)

        st.write("### Contact")
        st.write(f"üìß {email}")
        if github:
            st.write(f"[GitHub]({github})")
        if linkedin:
            st.write(f"[LinkedIn]({linkedin})")
```
